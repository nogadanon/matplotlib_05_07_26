```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve
```

Task 1  
Draw the lines for:   
x + y = 12    
x - y = 4  
Add grid, legend, and axis lines  
  
Task 2  
Use symbols, Eq, and solve to find the intersection point. Scatter the point and annotate it on the graph  
  
Task 3  
The label points are Y = np.array([2, 4, 8, 20])  
In the graph, the predicted Y values were Y_pred = np.array([2, 5, 10, 25])  
Calculate the MSE

```python
x_values = np.linspace(-5, 30, 10)
y_values1 = -x_values + 12
y_values2 = (x_values) - 4

plt.axhline(0, color='black', linewidth=0.5, linestyle='-')
plt.axvline(0, color='black', linewidth=0.5, linestyle='-')
plt.grid(color='gray', linestyle='--', linewidth=0.2)
plt.plot(x_values, y_values1, color='red', label='y = -x + 12')
plt.plot(x_values, y_values2, color='blue', label='y = x - 4')
plt.title('y = -x + 12 = x - 4')
plt.legend()

# Task 2
x = symbols('x')
eq1 = -x + 12
eq2 = x - 4
solution = solve(Eq(eq1, eq2), x)

x_junc = solution[0]
y_junc = x_junc - 4
plt.scatter(x_junc, y_junc, color='green', zorder=5)
plt.annotate(f"({x_junc}, {y_junc})", xy=(x_junc, y_junc), xytext=(x_junc +1, y_junc -0.5), fontsize=10, color='green')

# Task 3
Y = np.array([2, 4, 8, 20])
Y_pred = np.array([2, 5, 10, 25])
mse = ((Y - Y_pred)**2).mean()
mse

```
